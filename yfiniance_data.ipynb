```python
from datetime import datetime
import pandas as pd

date_time_str = '2021-09-25'

date_time_obj = datetime.fromisoformat(date_time_str)
date_time_obj
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import Normalizer
import numpy as np

times = ["2021-12-25","2021-09-25","2021-06-26","2021-03-27"]
times = list(map(datetime.fromisoformat, times))
times.reverse()

transformer = Normalizer(norm="l2").fit(apple.quarterly_balance_sheet)

plt.figure(figsize=(20,10))
for index , row in apple.quarterly_balance_sheet.iterrows():

    print(transformer.transform(row.values.reshape(1, -1)))
    plt.scatter(times, transformer.transform(row.values.reshape(1, -1))[0] , label=row.name)

plt.legend(loc="best")
plt.show()
```

```python
#now we can try to normalize by total current assets
apple_fundementals = apple.quarterly_balance_sheet

for col in apple_fundementals.columns:
    apple_fundementals[col] /= apple_fundementals[col]["Total Current Assets"]

plt.figure(figsize=(20,10))
for index , row in apple_fundementals.iterrows():
    plt.scatter(times, row.values, label=row.name)

plt.legend(loc="best")
plt.show()
```
